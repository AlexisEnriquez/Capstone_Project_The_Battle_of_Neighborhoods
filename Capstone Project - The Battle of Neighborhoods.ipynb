```python
# run the Nearby Venues function and print outputs
monterrey_venues = getNearbyVenues(names=df['Municipio'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                                  )
```

```python
# One hot encoding
monterrey_onehot = pd.get_dummies(monterrey_venues[['Venue Category']], prefix="", prefix_sep="")

# add municipality column back to dataframe
monterrey_onehot['Municipio'] = monterrey_venues['Municipio'] 

# move municipality column to the first column
fixed_columns = [monterrey_onehot.columns[-1]] + list(monterrey_onehot.columns[:-1])
monterrey_onehot = monterrey_onehot[fixed_columns]

monterrey_onehot.head()
```

```python
#Next, let's group rows by municipality and by taking the mean of the frequency of occurrence of each category

monterrey_grouped = monterrey_onehot.groupby('Municipio').mean().reset_index()
monterrey_grouped
```

```python
#Let's print each municipality along with the top 5 most common venues
num_top_venues = 5

for hood in monterrey_grouped['Municipio']:
    print("----"+hood+"----")
    temp = monterrey_grouped[monterrey_grouped['Municipio'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Municipio']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
county_venues_sorted = pd.DataFrame(columns=columns)
county_venues_sorted['Municipio'] = monterrey_grouped['Municipio']

for ind in np.arange(monterrey_grouped.shape[0]):
    county_venues_sorted.iloc[ind, 1:] = return_most_common_venues(monterrey_grouped.iloc[ind, :], num_top_venues)

county_venues_sorted.head()
```

```python
# set number of clusters
kclusters = 5

monterrey_grouped_clustering = monterrey_grouped.drop('Municipio', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(monterrey_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
county_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

monterrey_merged = df

monterrey_merged = monterrey_merged.join(county_venues_sorted.set_index('Municipio'), on='Municipio')

monterrey_merged.head() # check the last columns!
```
